## load the data

```python
import pandas as pd
data=pd.read_csv('L2D2.csv',delimiter=',',dtype='object',header=None)
data
```

```python
data[60].replace(['R','M'],['1','0'],inplace=True)
data[60]
```

```python
data=data.apply(pd.to_numeric,errors='coerce').fillna(0)
data.dtypes
```

```python
labels=['M','R']
labels
```

```python
data
```

## Split data

```python
X=data.drop(60,axis=1)
y=data[60]
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(X,y,test_size=0.33,random_state=15)
```

## KerasClassifier Model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import numpy


seed = 7
numpy.random.seed(seed)

def create_model():
    model = Sequential()
    model.add(Dense(12, input_dim=60, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
    
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
model1 = KerasClassifier(build_fn=create_model, epochs=15, batch_size=10)
results = cross_val_score(model1, x_tr, y_tr, cv=kfold) 
```

```python
score_1=print(results.mean()*100)
score_1
```

# Question 2: Check whether accuracy will improve with the larger network

```python
numpy.random.seed(15)

model2=Sequential()

model2.add(Dense(32, input_dim=60, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(22, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(12, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model2.fit(x_tr,y_tr,epochs=150)
```

```python
model2.evaluate(x_te,y_te)
```

# Question 3: Check whether accuracy will be improved after Dropout or not

```python
from keras.layers import Dropout

numpy.random.seed(15)


model3=Sequential()

model3.add(Dense(32, input_dim=60, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.6,seed=15))
model3.add(Dense(22, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.6,seed=15))
model3.add(Dense(12, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.6,seed=15))
model3.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model3.fit(x_tr,y_tr,epochs=150)
```

```python
model3.evaluate(x_te,y_te)
```

Question 4: Check whether accuracy will be improved after or not with delay learning rate; use sgd optimizer

```python
from keras.layers import Dropout
import numpy

numpy.random.seed(15)


model4=Sequential()
model4.add(Dropout(0.2, input_shape=(60,)))
model4.add(Dense(256, input_dim=60, kernel_initializer='uniform', activation='relu'))
model4.add(Dropout(0.6,seed=15))
model4.add(Dense(192, kernel_initializer='uniform', activation='relu'))
model4.add(Dropout(0.6,seed=15))
model4.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model4.add(Dropout(0.6,seed=15))
model4.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

model4.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model4.fit(x_tr,y_tr,epochs=150)
```

```python
model4.evaluate(x_te,y_te)
```
